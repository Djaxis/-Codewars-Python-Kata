# Execute the code below

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
link = "https://odre.opendatasoft.com/explore/dataset/temperature-quotidienne-departementale/download?format=csv&timezone=Europe/Berlin&use_labels_for_header=false"
# Charger les données dans un DataFrame
energie = pd.read_csv(link, delimiter=";")

# Trier le DataFrame en utilisant la colonne "code_insee_departement"
df_sorted = energie.sort_values(by="code_insee_departement")
df_sorted
```

### ECLAIRCIR ET NETOYER DE LA DATA

```
### ECLAIRCIR ET NETOYER DE LA DATA
# Supprimez les lignes avec des données manquantes si nécessaire
df_sorted.dropna(inplace=True)

# Tri du DataFrame dans l'ordre croissant par code insee puis date
datatemp = df_sorted.sort_values(by=['code_insee_departement', 'date_obs'])

# Conversion de la colonne 'date_obs' en format de date
datatemp['date_obs'] = pd.to_datetime(datatemp['date_obs'])

# Séparation de la colonne 'date_obs' en trois colonnes distinctes
datatemp['jour'] = datatemp['date_obs'].dt.day
datatemp['mois'] = datatemp['date_obs'].dt.month
datatemp['annee'] = datatemp['date_obs'].dt.year

# Organiser les colonnes dans le bon ordre
column_order = ["code_insee_departement","departement","date_obs","mois","annee","jour","tmin", "tmax", "tmoy"]


# Tri du DataFrame dans l'ordre croissant par code insee puis date
datatemp = datatemp.sort_values(by=['code_insee_departement','date_obs',"annee","mois","jour"])

print('DATATEMP COLONNES')
print()
print(datatemp.columns)
print()

# Afficher le DataFrame
print(datatemp.head(40))

# Sauvegarde du DataFrame trié dans un fichier CSV
#df_sorted.to_csv('donnees_triees.csv', index=False)
```

##La moyenne des température MENSUELLE par code_insee_departement

```
##CREATION D'UN AUTRE TABLEAU MOYENNE MENSUELLE A##
# Conversion de la colonne 'date_obs' en format de date
datatemp['date_obs'] = pd.to_datetime(datatemp['date_obs'])

# Séparation de la colonne 'date_obs' en trois colonnes distinctes
datatemp['jour'] = datatemp['date_obs'].dt.day
datatemp['mois'] = datatemp['date_obs'].dt.month
datatemp['annee'] = datatemp['date_obs'].dt.year
# Groupement par code_insee_departement, mois et année, et calcul des moyennes
df_groupedMyoMensuel = datatemp.groupby(['code_insee_departement', 'mois', 'annee'])[['tmin', 'tmax', 'tmoy']].mean().reset_index()

# Renommage des colonnes
df_groupedMyoMensuel.rename(columns={'tmin': 'TminMoyMens', 'tmax': 'TmaxMoyMens', 'tmoy': 'TMoyMens'}, inplace=True)


# Tri du DataFrame dans l'ordre croissant par code insee puis date
df_groupedMyoMensuel = df_groupedMyoMensuel.sort_values(by=['code_insee_departement','annee', 'mois'])

print('df_groupedMyoMensuel COLONNES')
print()
print(df_groupedMyoMensuel.columns)
print()
# Afficher le DataFrame
print(df_groupedMyoMensuel.head(40))
```

##La moyenne ANNUELLE des température par code_insee_departement

```
##CREATION D'UN AUTRE TABLEAU MOYENNE ANNUELLE##

# Groupement par code_insee_departement année, et calcul des moyennes
df_grouped = datatemp.groupby(['code_insee_departement','annee'])[['tmin', 'tmax', 'tmoy']].mean().reset_index()

# Renommage des colonnes
df_grouped.rename(columns={'tmin': 'TminMoyAn', 'tmax': 'TmaxMoyAn', 'tmoy': 'TMoyAn'}, inplace=True)

# Tri du DataFrame dans l'ordre croissant par code insee puis date
df_groupedMyoAnnee = df_grouped.sort_values(by=['code_insee_departement','annee'])

# Afficher le DataFrame
print(df_groupedMyoAnnee)
# Afficher les 40 premières lignes du DataFrame df_groupedMyoAnnee
#print(df_groupedMyoAnnee.head(40))
```

```
##CREATION D'UN AUTRE TABLEAU MOYENNE TOUTES ANNEES CONFONDUS##

# Groupement par code_insee_departement année, et calcul des moyennes
df_grouped = datatemp.groupby(['code_insee_departement'])[['tmin', 'tmax', 'tmoy']].mean().reset_index()

# Renommage des colonnes
df_grouped.rename(columns={'tmin': 'MoyMinAll', 'tmax': 'MoyMaxAll', 'tmoy': 'MoyAllMoy'}, inplace=True)

# Tri du DataFrame dans l'ordre croissant par code insee puis date
df_groupedAnneeTotaleDept= df_grouped.sort_values(by=['code_insee_departement'])

# Afficher le DataFrame
print(df_groupedAnneeTotaleDept)

# Afficher les 40 premières lignes du DataFrame df_groupedMyoAnnee
#print(df_groupedMyoAnnee.head(40))
```
