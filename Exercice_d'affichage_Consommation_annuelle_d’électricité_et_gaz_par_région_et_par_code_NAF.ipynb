```python
import pandas as pd
from IPython.display import display

url = "https://www.data.gouv.fr/fr/datasets/r/d33eabc9-e2fd-4787-83e5-a5fcfb5af66d"

data = pd.read_csv(url, delimiter=";")
display(data)

```

```python
# Nombre de lignes
num_rows = len(data)

# Nombre de colonnes
num_columns = len(data.columns)

# Titres des colonnes
column_titles = data.columns.tolist()

print("Nombre de lignes :", num_rows)
print("Nombre de colonnes :", num_columns)
print("Titres des colonnes :", column_titles)
```

### **Description du fichier**
Opérateur: operateur[text] Nom de l’opérateur, qu’il soit distributeur ou transporteur

Année: annee[date] Année

Filière: filiere[text] Énergie concernée (Electricité ou Gaz)

Code catégorie consommation: code_categorie_consommation[text] La variable peut prendre 5 modalités:

RES: correspond aux consommations résidentielles ;

PRO: correspond aux consommation des petits professionnels (clients non résidentiels avec une puissance souscrite <=36 kva ) ;

ENT: correspond aux consommation des entreprises (clients non résidentiels avec une puissance souscrite >36 kva );

ENT-PRO: correspond aux consommations des entreprises et des petits professionnels, dans les cas où il n’a pas été possible de les dissocier;

RES-PRO: correspond aux consommations résidentielles et des petits professionnels, dans les cas où il n’a pas été possible de les dissocier

Libellé catégorie consommation: libelle_categorie_consommation[text] Libellé associé au code de la catégorie de consommation

Code Grand secteur: code_grand_secteur[text] Code du grand secteur d’activité

Libellé Grand Secteur: libelle_grand_secteur[text] Libellé du grand secteur d’activité

Code NAF: code_naf[text] Code NAF à 2 positions du secteur (NAF rev 2 2008)

Libellé du secteur NAF: libelle_secteur_naf2[text] Libellé Secteur NAF

Consommation (MWh): conso[double] Consommation (MWh). Nombre positif ou nul.

Nombre de points: pdl[int] Nombre de points de livraison. Nombre entier, positif ou nul.

Indice qualité: indqual[double] Pourcentage de la consommation annuelle est mesurée du 1er janvier au 31 décembre. Il est compris entre 0 et 1.

Nombre de mailles secretisées: nombre_mailles_secretisees[int] La secretisation des données vise a proteger les données personnelles (voir note methodologique). Si la maille est secretisée la colonne “Nombre de mailles secretisées” prend la valeur 1. Les colonnes “Consommation” et “Nombre de points” prennent la valeur 0.

Code Région: code_region[text] Code Région

Libellé Région: libelle_region[text] Libellé Région
